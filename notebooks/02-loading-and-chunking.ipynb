```python
import sys
from pathlib import Path

sys.path.insert(0,str(Path.cwd().parent))

from src.loading import load_vault
from src.chunking import _split_string_by_headers, Chunk


VAULT_DIR = Path.cwd()
```

```python
vault = load_vault(VAULT_DIR)

chunks = []
index = 0
for item in vault:
    for headings, section in _split_string_by_headers(item.content):

        breadcrumbs = ' > '.join([item.name,*headings])
        chunks.append(
            Chunk(
                text = f"# {breadcrumbs}\n\n{section}\n",
                metadata={
                    "source":item.path,
                    "file":item.name,
                    "headings":headings,
                    "index":index
                }
                )
                      )
        index += 1

chunks
```
