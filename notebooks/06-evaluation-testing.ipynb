```python
from rag_build.querying import search

def retrieval_evaluation():
    """Uses evaluation set to rate the retrieval of the RAG Pipeline"""

    correct_sources = 0

    for test in EVALUATION_SET:
        question, answer = list(test.items())[0]

        responses = search(question)
        sources = [response['metadata']['source'] for response in responses]
        correct_sources += int(answer in sources)

    
    return correct_sources/len(EVALUATION_SET)

        
retrieval_evaluation()

```
